**Q. What is a random variable in probability theory?**

Ans:

  - A random variable is a numerical value that represents the outcome of a random experiment.
  - It assigns a number to each possible outcome of an event.
  - Examples:
    - Tossing a coin → Heads = 1, Tails = 0
    - Rolling a die → random variable can be 1–6
    - Daily sales amount in a shop

**Q. What are the types of random variables?**

Ans:

  - There are two main types:

    1.  Discrete Random Variable
      - Takes countable values
      - Example: number of students in a class, number of heads in coin toss

    2. Continuous Random Variable
      - Takes infinite values within a range
      - Example: height, weight, temperature, time

**Q. What is a binomial distribution, and how is it used in probability?**

Ans:

  - A binomial distribution shows the probability of a fixed number of successes in a fixed number of independent trials.
  - Conditions:
    - Only two outcomes (success/failure)
    - Fixed number of trials
    - Same probability each trial
    - Independent trials

  - Example:
    - Tossing a coin 5 times → probability of getting exactly 3 heads.

  - Formula:
    - P(X=k) = C(n,k) p^k (1-p)^(n-k)

  - Used in:
    - Quality testing
    - Marketing success rate
    - Survey analysis

**Q. What is the standard normal distribution, and why is it important?**

Ans:

  - A standard normal distribution is a normal distribution with:
    - Mean = 0
    - Standard deviation = 1

    It is bell-shaped and symmetric.

  - Importance:
    - Used to calculate probabilities using Z-scores
    - Helps compare different datasets
    - Foundation of hypothesis testing

**Q. What is the Central Limit Theorem (CLT), and why is it critical in statistics?**

Ans:

  - The Central Limit Theorem states:
    
    For a large sample size, the sampling distribution of the mean becomes approximately normal, regardless of the population distribution.

  - Why important:
    - Allows use of normal distribution
    - Helps estimate population mean
    - Used in confidence intervals & hypothesis testing
    - Foundation of inferential statistics

**Q. What is the significance of confidence intervals in statistical analysis?**

Ans:

  - A confidence interval (CI) is a range of values used to estimate a population parameter.

  - Example:
    - 95% CI means we are 95% confident the true population mean lies within that range.

  - Significance:
    - Shows reliability of estimate
    - Helps decision making
    - Used in research & business analysis

**Q. What is the concept of expected value in a probability distribution?**

Ans:

  - The expected value (mean) is the average outcome of a random variable over many trials.

  - Formula:
    - E(X) = Σ [x × P(x)]

  - Example:
    - Rolling a fair die
    - Expected value = 3.5

  - Importance:
    - Predicts long-term average
    - Used in finance, risk analysis, games

Q. Write a Python program to generate 1000 random numbers from a normal
distribution with mean = 50 and standard deviation = 5. **bold text**

**Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.**  

(Include your Python code and output in the code box below.)


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random numbers
data = np.random.normal(loc=50, scale=5, size=1000)

# Compute mean and standard deviation
mean_val = np.mean(data)
std_val = np.std(data)

print("Mean:", mean_val)
print("Standard Deviation:", std_val)

# Histogram
plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution (mean=50, std=5)")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

**Q. You are working as a data analyst for a retail company.** The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

**● Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.**

**● Write the Python code to compute the mean sales and its confidence interval.**

(Include your Python code and output in the code box below.)

Ans:

  - How CLT helps:
    - Daily sales = sample data
    - CLT allows estimating population mean
    - Calculate mean and standard error
    - Build 95% confidence interval

  - Formula:
    - CI = mean ± Z * (σ/√n)

      Where Z = 1.96 for 95%

```python
import numpy as np
import scipy.stats as stats

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Mean and std
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)
n = len(daily_sales)

# Standard error
se = std_sales / np.sqrt(n)

# 95% CI
z = 1.96
lower = mean_sales - z * se
upper = mean_sales + z * se

print("Mean Daily Sales:", mean_sales)
print("95% Confidence Interval:", (lower, upper))
```
