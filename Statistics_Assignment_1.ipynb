**Q. What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

  - Descriptive statistics summarize and describe the features of a dataset.
  - They help understand what the data shows.

  - Examples:
    - Mean marks of students in a class
    - Median salary of employees
    - Bar chart of monthly sales
    
  - Inferential statistics use sample data to make predictions or conclusions about a larger population.

  - Examples:
    - Predicting election results using a sample survey
    - Estimating average income of a city using sample households
    - Hypothesis testing (e.g., does a new drug work?)

  - Difference:
    - Descriptive → describes existing data.
    - Inferential → draws conclusions or predictions from data

**Q. What is sampling in statistics? Explain the differences between random and stratified sampling.**

  - Sampling is the process of selecting a subset (sample) from a population to study and analyze.

  - Random Sampling:
    - Every individual has an equal chance of being selected.
    
     Example:
      - Selecting 50 students randomly from a college using a random number generator.

  - Stratified Sampling:
    - Population is divided into groups (strata) and samples are taken from each group.

    Example:
      - Dividing students into departments (science, commerce, arts) and selecting samples from each.

**Q. Define mean, median, and mode. Explain why these measures of central tendency are important.**

  - Mean: Average of all values

    Mean = sum of values / total values
  - Median: Middle value when data is arranged
  - Mode: Most frequently occurring value

  - Importance:
    - Help understand central value of data
    - Useful in comparison
    - Helps detect skewness and distribution
    - Used in business, economics, research

**Q. Explain skewness and kurtosis. What does a positive skew imply about the data?**

  - Skewness: Measures asymmetry of data distribution.
    - Positive skew → tail on right side
    - Negative skew → tail on left

  - Positive skew implies:
    - Mean > Median
    - Most values are smaller
    - Few very large values pull mean right

  - Kurtosis: Measures “peakedness” of distribution.
    - High kurtosis → more outliers, sharp peak
    - Low kurtosis → flat distribution

**Q. Implement a Python program to compute the mean, median, and mode of a given list of numbers.**

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

(Include your Python code and output in the code box below.)

```python
import statistics

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

mean_val = statistics.mean(numbers)
median_val = statistics.median(numbers)
mode_val = statistics.multimode(numbers)

print("Mean:", mean_val)
print("Median:", median_val)
print("Mode:", mode_val)
```

**Q. Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:**

list_x = [10, 20, 30, 40, 50]

list_y = [15, 25, 35, 45, 60]

(Include your Python code and output in the code box below.)

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

covariance = np.cov(list_x, list_y)[0][1]
correlation = np.corrcoef(list_x, list_y)[0][1]

print("Covariance:", covariance)
print("Correlation coefficient:", correlation)
```

**Q. Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:**

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

(Include your Python code and output in the code box below.)

```python
import matplotlib.pyplot as plt
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

plt.boxplot(data)
plt.title("Boxplot of Data")
plt.show()

# Finding outliers using IQR
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = [x for x in data if x < lower_bound or x > upper_bound]

print("Outliers:", outliers)
```

**Q. You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.**

**● Explain how you would use covariance and correlation to explore this relationship.**

**● Write Python code to compute the correlation between the two lists:**

advertising_spend = [200, 250, 300, 400, 500]

daily_sales = [2200, 2450, 2750, 3200, 4000]

(Include your Python code and output in the code box below.)


Ans:
  - How covariance & correlation help:
    - Covariance: Shows direction of relationship
    - Correlation: Shows strength (−1 to +1)
    - Positive value → as advertising increases, sales increase

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

correlation = np.corrcoef(advertising_spend, daily_sales)[0][1]

print("Correlation between advertising and sales:", correlation)
```

**Q. Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.**

**● Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.**

**● Write Python code to create a histogram using Matplotlib for the survey data:**

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

(Include your Python code and output in the code box below.)

Ans:

  - Useful summary statistics:
    - Mean → average satisfaction
    - Median → middle satisfaction
    - Standard deviation → variability
    - Min & Max → range

  - Visualizations:
    - Histogram → distribution
    - Boxplot → outliers
    - Bar chart → frequency

```python
import matplotlib.pyplot as plt

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

plt.hist(survey_scores, bins=6, edgecolor='black')
plt.title("Customer Satisfaction Histogram")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.show()
```
